```python
%matplotlib widget

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
import pandas as pd
# %matplotlib notebook
# %matplotlib ipympl
```

```python
def readData(network_size, seed):
    directory = "../data/TSP/N{:d}-{:d}/".format(network_size, seed)
    pos = pd.read_csv(directory + "position.txt", header=None).to_numpy(dtype=np.float16)
    data = pd.read_csv(directory + "data.txt", header=None).to_numpy(dtype=np.float16)

    return pos.transpose(), data[:, :network_size].astype(np.int16), data[:, -2], data[:, -1]

```

```python
# Get data
network_size = 20
seed = 0
position, order, temperature, energy = readData(network_size=network_size, seed=seed)
x_pos = position[0]
y_pos = position[1]
```

```python

fig, ax = plt.subplots(figsize=(10,10))
fig.patch.set_facecolor('white')

def update(frame):
    ax.clear()
    for node in range(network_size):
        next = order[frame][node]
        # print(x_pos[node], y_pos[node], x_pos[next], y_pos[next])
        ax.plot([x_pos[node], x_pos[next]], [y_pos[node], y_pos[next]], linewidth=2, color='red')
        ax.text(x_pos[node], y_pos[node], str(node), fontsize=20)
    ax.plot(position[0], position[1], marker='.', color='black', linestyle='None', markersize=20, label="T={:.2f}, E={:.6f}".format(temperature[frame], energy[frame]))
    ax.set_xlim(-1,1)
    ax.set_ylim(-1,1)
    ax.text(0.8, 0.9, "Frame: {:d}".format(frame), fontsize=20)
    ax.legend(loc="lower right", fontsize=20)


ani = animation.FuncAnimation(fig=fig, func=update, frames=len(order), blit=False, interval=100, repeat=False)
# ani.save("../data/TSD/N{:d}-{:d}/".format(network_size, seed) + "figure.gif", writer='ffmpeg', dpi=100)

plt.show()
```
